```python
# Define the buy date for price analysis
buy_date = pd.Timestamp('2025-03-12')
start = "2025-01-01"

# Download Thales stock data from Euronext Paris (Ticker: HO.PA) https://finance.yahoo.com/quote/HO.PA/
# Time period: From January 1, 2022 to the current date
data = yf.download("HO.PA", start, end=pd.Timestamp.now().strftime('%Y-%m-%d'))

# Adjust data structure: Reset index to have 'Date' as a column
# and then set it back as index for better data queries
data = data.reset_index()
data.set_index('Date', inplace=True)

# Determine maximum price and corresponding date
# Check if these variables already exist to avoid overwriting
if 'max_date' not in locals() or 'max_price' not in locals():
    max_price = data[('Close', 'HO.PA')].max()  # Highest closing price
    max_date = data.loc[data[('Close', 'HO.PA')] == max_price].index[0]  # Date of highest price

# Get purchase price on the defined buy date
buy_price = data.loc[buy_date, ('Close', 'HO.PA')]

# Get current price (last entry in the data)
current_price = data[('Close', 'HO.PA')].iloc[-1]
current_date = data.index[-1]

# Create visualization with size 10x5 inches
plt.figure(figsize=(10, 5))

# Main line: Draw stock price over time
plt.plot(data[('Close', 'HO.PA')], label='Thales Stock Price')

# Mark the highest point with a red dot and annotation
plt.plot(max_date, max_price, 'ro', label='Highest Point')
plt.annotate(f'{max_price:.2f} EUR', xy=(max_date, max_price), xytext=(max_date, max_price + 5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Mark the purchase date with a yellow dot and annotation
plt.plot(buy_date, buy_price, 'yo', label='Purchase Date')
plt.annotate(f'{buy_price:.2f} EUR', xy=(buy_date, buy_price), 
             xytext=(buy_date, buy_price + 5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Mark the current price with a green dot and annotation
plt.plot(current_date, current_price, 'go', label='Current Price')
plt.annotate(f'{current_price:.2f} EUR', xy=(current_date, current_price), 
             xytext=(current_date, current_price - 10),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Add chart details
plt.title('Thales Stock Price (2022-2025)')
plt.xlabel('Date')
plt.ylabel('Price in EUR')
plt.legend()
plt.grid(True)
```

```python
# Calculate profit/loss
profit_loss = current_price - buy_price
profit_loss_percent = (profit_loss / buy_price) * 100

# Create extended table with price overview and profit/loss
price_table = pd.DataFrame({
    'Description': ['Purchase Price', 'Current Price', 'Maximum Price', 'Profit/Loss', 'Profit/Loss (%)'],    
    'Value': [
        f"{buy_price:.2f} EUR", 
        f"{current_price:.2f} EUR", 
        f"{max_price:.2f} EUR",
        f"{profit_loss:.2f} EUR",
        f"{profit_loss_percent:.2f}%"
    ]
})

print("\nPrice Overview and Profit/Loss:")
print(price_table)

# Print result
if profit_loss > 0:
    print(f"\nProfit: {profit_loss:.2f} EUR ({profit_loss_percent:.2f}%)")
else:
    print(f"\nLoss: {profit_loss:.2f} EUR ({profit_loss_percent:.2f}%)")

# Calculate price difference between highest price and purchase price
price_difference_scalar = max_price - buy_price
```
